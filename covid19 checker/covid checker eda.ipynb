```python
countries = df['Country'].unique()
countries
```

```python
## Country wise severity distribution
index = 0
row=1
for country in countries:
    index = index + 1
    val1 = len(df[(df['Country']==country) & (df['Severity_Severe']==1)])
    val2 = len(df[(df['Country']==country) & (df['Severity_Moderate']==1)])
    val3 = len(df[(df['Country']==country) & (df['Severity_Mild']==1)])
    val4 = len(df[(df['Country']==country) & (df['Severity_None']==1)])
    val = [val1, val2, val3, val4]
    plt.figure(figsize=(20, 10))
    plt.subplot(row,1,index)
    plt.bar(severity_values, val, width=0.4)
    plt.xlabel('Severity distribution')
    plt.ylabel('Number of people affected')
    plt.title(f'{country} cases distribution')
    row = row + 1
    
```

```python
## Country wise age distribution
index = 0
row=1
for country in countries:
    index = index + 1
    val1 = len(df[(df['Country']==country) & (df['Age_0-9']==1)])
    val2 = len(df[(df['Country']==country) & (df['Age_10-19']==1)])
    val3 = len(df[(df['Country']==country) & (df['Age_20-24']==1)])
    val4 = len(df[(df['Country']==country) & (df['Age_25-59']==1)])
    val5 = len(df[(df['Country']==country) & (df['Age_60+']==1)])
    val = [val1, val2, val3, val4, val5]
    plt.figure(figsize=(20, 10))
    plt.subplot(row,1,index)
    plt.bar(age_values, val, width=0.4)
    plt.xlabel('Age Distribution')
    plt.ylabel('Number of people affected')
    plt.title(f'{country} cases distribution')
    row = row + 1
```

```python
## Country gender age distribution
index = 0
row=1
for country in countries:
    index = index + 1
    val1 = len(df[(df['Country']==country) & (df['Gender_Female']==1)])
    val2 = len(df[(df['Country']==country) & (df['Gender_Male']==1)])
    val3 = len(df[(df['Country']==country) & (df['Gender_Transgender']==1)])
    val = [val1, val2, val3]
    plt.figure(figsize=(20, 10))
    plt.subplot(row,1,index)
    plt.bar(['Females', 'Males', 'Trans'], val, width=0.4)
    plt.xlabel('Gender Distribution')
    plt.ylabel('Number of people affected')
    plt.title(f'{country} cases distribution')
    row = row + 1
```
