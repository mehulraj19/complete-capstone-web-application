```python
confirmed_country = confirmed_global_df[confirmed_global_df['Country/Region']==country].reset_index(drop=True)
recovered_country = recovered_global_df[recovered_global_df['Country/Region']==country].reset_index(drop=True)
deaths_country = deaths_global_df[deaths_global_df['Country/Region']==country].reset_index(drop=True)
```

```python
df = confirmed_country.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=confirmed_global_df.columns[4:], var_name='Date', value_name='ConfirmedCases')
confirmed_df = df.groupby(["Date", "Country/Region"])[['Date', 'Country/Region', 'ConfirmedCases']].sum().reset_index()
confirmed_df.columns = ['Date', 'Country', 'Confirmed']
confirmed_df.head()
```

```python
df = recovered_country.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=recovered_global_df.columns[4:], var_name='Date', value_name='RecoveredCases')
recovered_df = df.groupby(["Date", "Country/Region"])[['Date', 'Country/Region', 'RecoveredCases']].sum().reset_index()
recovered_df.columns = ['Date', 'Country', 'Recovered']
recovered_df.head()
```

```python
df = deaths_country.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=deaths_global_df.columns[4:], var_name='Date', value_name='DeathCases')
deaths_df = df.groupby(["Date", "Country/Region"])[['Date', 'Country/Region', 'DeathCases']].sum().reset_index()
deaths_df.columns = ['Date', 'Country', 'Deaths']
deaths_df.head()
```

## Country wise

```python
df = confirmed_global_df.melt(id_vars=['Province/State', 'Country/Region', 'Lat', 'Long'], value_vars=confirmed_global_df.columns[4:], var_name='Date', value_name='ConfirmedCases')
df1 = df.groupby(["Date", "Country/Region"])[['Date', 'Country/Region', 'ConfirmedCases']].sum().reset_index()
df1.columns = ['Date', 'Country', 'Confirmed']
df1.head()
```

```python
df1.set_index('Date', inplace=True)
df1.sort_values(by=['Confirmed'], ascending=False, inplace=True)
df1.head(50)
```

```python
arr = df1.Country.unique()
```

```python
arr
```

```python
len(arr)
```

```python
country_wise_data = {}
for i in range(len(arr)):
    val = df1.loc[df1['Country']==arr[i], 'Confirmed'].sum()
    country_wise_data[arr[i]] = val
country_wise_data
```

```python
country_wise_data = sorted(country_wise_data.items(), key=lambda x: x[1])
country_wise_data
```

## Top 10 countries having max confirmed cases

```python
val = len(country_wise_data)
val
```

```python
countries = {}
for i in range(val-1, val-11, -1):
    countries[country_wise_data[i][0]] = country_wise_data[i][1]
countries
```

```python
plt1 = plt.figure(figsize=(20, 10))
plt.bar(list(countries.keys()), list(countries.values()), width=0.4)
plt.xlabel('Countries')
plt.ylabel('Confirmed Cases')
plt.title('Majorly Affected Countries based on Confirmed Cases')
plt.show()
```
